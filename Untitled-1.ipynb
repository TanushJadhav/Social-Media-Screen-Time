```python
for i in df:
    if df.dtypes[i]=='object':
        print(f"Unique values of {i} are: {list(df[i].unique())}")
```

```python
df['gender'] = df['gender'].replace(['male', 'female', 'non-binary'], [1, 2, 3])
df['platform'] = df['platform'].replace(['Instagram', 'Facebook', 'YouTube'], [1, 2, 3])
df['interests'] = df['interests'].replace(['Sports', 'Travel', 'Lifestlye'], [1, 2, 3])
df['location'] = df['location'].replace(['United Kingdom', 'Australia', 'United States'], [1, 2, 3])
df['demographics'] = df['demographics'].replace(['Urban', 'Sub_Urban', 'Rural'], [1, 2, 3])
df['profession'] = df['profession'].replace(['Software Engineer', 'Student', 'Marketer Manager'], [1, 2, 3])
```
